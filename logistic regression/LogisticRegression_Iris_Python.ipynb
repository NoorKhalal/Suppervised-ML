## Logistic regression with one feature

```python
print("Features")
print(iris.feature_names)

print("\nData head")
print(iris.data[95:105,:])

print("\nClass head")
print(iris.target[95:105])

print("\nClass names")
print(iris.target_names)
```

```python
iris.target
```

```python
iris["target"][95:105]
```

```python
# Let's try with a single predictor, the 'petal width'
X = iris["data"][:, 3:]
X
```

```python
# We get the target values for one type of flowers
y = (iris["target"] == 2).astype(int) # 1 if Iris-Virginica, else 0
y
```

```python
# Prepare the Logistic Regression model and train it
log_reg = LogisticRegression()
log_reg.fit(X, y)
?LogisticRegression
```

```python
log_reg.intercept_
```

```python
log_reg.score
```

```python
# Let's look at the probabilities from the model
# Reminder: we have petal widths from 0 to 3 cm

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
#https://stackoverflow.com/questions/18691084/what-does-1-mean-in-numpy-reshape
print(X_new.shape)
y_proba = log_reg.predict_proba(X_new)
print(y_proba)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris-Virginica");
```

```python
# The figure proposed below offers a better visibility

X_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_proba = log_reg.predict_proba(X_new)
# The decision boundary is considered at 0.5
decision_boundary = X_new[y_proba[:, 1] >= 0.5][0]

plt.figure(figsize=(8, 3))
plt.plot(X[y==0], y[y==0], "bs")
plt.plot(X[y==1], y[y==1], "g^")
plt.plot([decision_boundary, decision_boundary], [-1, 2], "k:", linewidth=2)
plt.plot(X_new, y_proba[:, 1], "g-", linewidth=2, label="Iris-Virginica")
plt.plot(X_new, y_proba[:, 0], "b--", linewidth=2, label="Not Iris-Virginica")
plt.text(decision_boundary+0.02, 0.15, "Decision  boundary", fontsize=14, color="k", ha="center")
plt.arrow(decision_boundary, 0.08, -0.3, 0, head_width=0.05, head_length=0.1, fc='b', ec='b')
plt.arrow(decision_boundary, 0.92, 0.3, 0, head_width=0.05, head_length=0.1, fc='g', ec='g')
plt.xlabel("Petal width (cm)", fontsize=14)
plt.ylabel("Probability", fontsize=14)
plt.legend(loc="center left", fontsize=14)
plt.axis([0, 3, -0.02, 1.02])
plt.show()
```

**Objective** The model is trained to estimate the probability that a new flower is an Iris-Virginica

*Triangles* : The petal width of Iris-Virginica flowers ranges from 1.4cm to 2.5cm<br>

*Squares* : The other iris flowers generally have smaller petal width, from 0.1cm to 1.8cm.<br>

There is a bit of overlap, with also highly confident areas at extremities.<br>
The decision boundary is at 1.6cm.

```python
# We predict directly the class with the predict function
log_reg.predict([[1.7], [1.5], [1.8]])
```

## Logistic regression with two features

```python
# We consider the petal length and the petal width as predictors
X = iris["data"][:, (2, 3)]  # petal length, petal width
y = (iris["target"] == 2).astype(int)
```

```python
# Prepare the Logistic Regression model and train it
#log_reg = LogisticRegression()
log_reg = LogisticRegression(C=10) #why?
log_reg.fit(X, y)

# NB:
# More info at https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
```

```python
# We generate an unseen dataset with plausible features values
x0, x1 = np.meshgrid(
        np.linspace(2.9, 7, 500).reshape(-1, 1),
        np.linspace(0.8, 2.7, 200).reshape(-1, 1),
    )
X_new = np.c_[x0.ravel(), x1.ravel()]
```

```python
# Predict the probabilities 
y_proba = log_reg.predict_proba(X_new)
y_proba
```

```python
X[y == 1]
```

```python
plt.figure(figsize=(10, 4))
plt.plot(X[y==0, 0], X[y==0, 1], "bs")
plt.plot(X[y==1, 0], X[y==1, 1], "g^")

zz = y_proba[:, 1].reshape(x0.shape)
contour = plt.contour(x0, x1, zz, cmap=plt.cm.brg)


left_right = np.array([2.9, 7])
boundary = -(log_reg.coef_[0][0] * left_right + log_reg.intercept_[0]) / log_reg.coef_[0][1]

plt.clabel(contour, inline=1, fontsize=12)
plt.plot(left_right, boundary, "k--", linewidth=3)
plt.text(3.5, 1.5, "Not Iris-Virginica", fontsize=14, color="b", ha="center")
plt.text(6.5, 2.3, "Iris-Virginica", fontsize=14, color="g", ha="center")
plt.xlabel("Petal length", fontsize=14)
plt.ylabel("Petal width", fontsize=14)
plt.axis([2.9, 7, 0.8, 2.7])
plt.show()
```

**Objective** The model has been trained to estimate the probability that a new flower is an Iris-Virginica

*Dashed line* : Decision boundary (50%).<br>

*Parallel line* : Points where the model outputs a specific probability (from 15% to 90%).<br>

All the flowers beyond the top-right line have an over 90% chance of being Iris-Virginica according to the model.

**C (float, default=1.0)**<br>

*Inverse* of regularization strength ($\alpha$); must be a positive float. Like in support vector machines, smaller values specify stronger regularization.
